```python
import os
from pydub import AudioSegment

def convert_to_mono(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".wav"):
                file_path = os.path.join(root, file)
                try:
                    # Load the stereo audio file
                    audio = AudioSegment.from_file(file_path)
                    
                    # Check if the file is already mono
                    if audio.channels > 1:
                        # Convert to mono
                        mono_audio = audio.set_channels(1)
                        
                        # Overwrite the original file
                        mono_audio.export(file_path, format="wav")
                        print(f"Converted to mono: {file_path}")
                    else:
                        print(f"Already mono: {file_path}")
                except Exception as e:
                    print(f"Failed to process {file_path}: {e}")

if __name__ == "__main__":
    library_path = "/Users/marek/Desktop/beatgrid/Samples"
    convert_to_mono(library_path)
```
